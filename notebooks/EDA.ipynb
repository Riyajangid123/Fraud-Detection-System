```python
percentage = df["Class"].value_counts(normalize=True) * 100
print(percentage)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Class", data=df)
plt.title("Class Distribution")
plt.show()
```

```python
corr = df.corr()["Class"].sort_values(ascending=False)

plt.figure(figsize=(6, 10))
sns.heatmap(corr.to_frame(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation with Target (Class)")
plt.show()
```

I analyzed feature correlation with the target variable. Features like V11 and V4 showed relatively higher correlation with fraud. Since features are PCA-transformed, I focused on correlation strength rather than direct interpretation.
